# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10, 10)) # (width ,heigth)

# 3. plot data
ax.plot(x,y)

# 4. customize plot
ax.set(title = "Simple plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Two options for subplots

```python
# Subplot option - 1
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows = 2,
                                           ncols = 2,
                                           figsize = (10, 5))

# plot toeach different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_price.keys(), nut_butter_price.values());
ax4.hist(np.random.rand(1000));
```

```python
# Subplot option - 2
fig, ax = plt.subplots(nrows = 2,
                      ncols = 2,
                      figsize = (10, 5))

# plot to each different index
c = ['red', 'yellow', 'black']
ax[0, 0].plot(x, x/2, color = 'red');
ax[0, 1].scatter(np.random.random(10), np.random.random(10), color = "violet");
ax[1, 0].bar(nut_butter_price.keys(), nut_butter_price.values(), color = c );
ax[1, 1].hist(np.random.rand(1000), color = "yellow", ec="orange", lw = 3);
```

### Which one should youuse?(pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use pyplot method
* When plotting something more advanced , use matplotlib OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# pyplot method
over_50.plot(kind = "scatter",
            x = "age",
            y ="chol",
            c = 'target');
```

```python
# OO methodwith pyplot
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind  = "scatter",
            x = "age",
            y = "chol",
            c = 'target',
            ax = ax);

#ax.set_xlim([45, 100]);
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

#plot the data
scatter = ax.scatter(x = over_50["age"],
                    y =  over_50["chol"],
                    c = over_50["target"])

#Customize the plot
ax.set(title = "Heart Dieseas and Choletrol Levels",
      xlabel = "Age",
       ylabel = "Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = "dashed",c = "r");
```

```python
over_50.head()
```

```python
# subplot of chol, age, thalch
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                     sharex = True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], 
                      y = over_50["chol"],
                     c = over_50["target"])

# Customize ax0
ax0.set(title = "Heart dieseas and cholestrol levels",
       xlabel = "Age",
       ylabel = "Cholestrol")

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title = "target")

# Add a meanline
ax0.axhline(y  = over_50["chol"].mean(),
           linestyle = "dashed",
           c = "r");


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"], 
                      y = over_50["thalach"],
                     c = over_50["target"])

# Customize ax1
ax1.set(title = "Heart dieseas and thalach levels",
       xlabel = "Age",
       ylabel = "Thalach")

#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title = "target")

# Add a meanline
ax1.axhline(y  = over_50["thalach"].mean(),
           linestyle = "dashed",
           c = "r");


#Add a title to figure
fig.suptitle("Heart Dieseas Analysis",fontsize = 16, fontweight = "bold");
```

```python
#Customize the y and x axis limitations




# subplot of chol, age, thalch
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize = (10, 10),
                     sharex = True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"], 
                      y = over_50["chol"],
                     c = over_50["target"],
                     cmap = "autumn")

# Customize ax0
ax0.set(title = "Heart dieseas and cholestrol levels",
       xlabel = "Age",
       ylabel = "Cholestrol")

# Change the X axis limit
ax0.set_xlim([50, 80])

#Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title = "target")

# Add a meanline
ax0.axhline(y  = over_50["chol"].mean(),
           linestyle = "dashed",
           c = "r");


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"], 
                      y = over_50["thalach"],
                     c = over_50["target"],
                     cmap = "winter")

# Customize ax1
ax1.set(title = "Heart dieseas and thalach levels",
       xlabel = "Age",
       ylabel = "Thalach")

#Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title = "target")

# Add a meanline
ax1.axhline(y  = over_50["thalach"].mean(),
           linestyle = "dashed",
           c = "r");


#Add a title to figure
fig.suptitle("Heart Dieseas Analysis",fontsize = 16, fontweight = "bold");
```
